```python
#Fixing columns types

df.info()

#columns to fix: 'distance', 'consume', 'temp_inside'
```

```python
#'distance'

df['temp_inside'].value_counts()
df['temp_inside'] = df['temp_inside'].str.replace(',','.')
df['temp_inside'] = df['temp_inside'].apply(lambda x: float(x))

df['consume'].value_counts()
df['consume'] = df['consume'].str.replace(',','.')
df['consume'] = df['consume'].apply(lambda x: float(x))

df['distance'].value_counts()
df['distance'] = df['distance'].str.replace(',','.')
df['distance'] = df['distance'].apply(lambda x: float(x))

```

```python
df.info() # values types fixed!
```

```python
# dealing with Nan in temp_inside

df['temp_inside'].value_counts(dropna=False)
df['temp_inside'].mean() #here the mean is very close to the mode, so we'll replace the Nan values for this

df['temp_inside'] = df['temp_inside'].fillna(df['temp_inside'].mean())

df['temp_inside'].isna().sum()
```
